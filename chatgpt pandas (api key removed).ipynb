```python
def reponse_to_text_list(response):
    text_list = []
    for s in response['choices'][0]['text'].split('\n'):
        if s:
            text_list.append(s)
    return text_list
```

```python
response = openai.Completion.create(
  model = "text-davinci-003",
  prompt = "列出10條數Data Scientist的面試問題",
  temperature = 0.5,
  max_tokens = 1024,
)
```

```python
ds_questions = reponse_to_text_list(response)
for q in ds_questions:
    print(q)
```
