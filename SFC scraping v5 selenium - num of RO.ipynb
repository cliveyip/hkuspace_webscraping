```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
import re
import pandas as pd
import time
```

```python
df = pd.read_excel('sfc_list.xlsx')
```

```python
cerefs = df['ceref'].to_list()
```

```python
len(cerefs)
```

```python
cerefs[:5]
```

```python
# get RO names

df_ro_names = pd.DataFrame()

options = Options()
options.add_argument('--headless=new')
driver = webdriver.Chrome(options=options)

for ref in cerefs:
    names = []
    url = "https://apps.sfc.hk/publicregWeb/corp/"+ref+"/ro"
    driver.get(url)
    print(url)
    time.sleep(2)
    
    name_grid = driver.find_elements(By.CLASS_NAME, "x-grid-cell-gridcolumn-1019")
    
    try:
        name_grid = driver.find_elements(By.CLASS_NAME, "x-grid-cell-gridcolumn-1019")
        for name in name_grid:
            names.append(name.text)
    except:
        names.append('no RO found')

    df1 = pd.DataFrame({'ceref': [ref]*len(names), 'ro_name': names})

    df_ro_names = pd.concat([df_ro_names, df1], ignore_index=True)

driver.quit()
```
