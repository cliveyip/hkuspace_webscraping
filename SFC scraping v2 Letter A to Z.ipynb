```python
def response_to_df(response):
    df = pd.DataFrame(response.json())
    items_expanded = df['items'].apply(pd.Series)
    df = pd.concat([df.drop('items', axis=1), items_expanded], axis=1)
    address_expanded = df['address'].apply(pd.Series)
    df = pd.concat([df.drop('address', axis=1), address_expanded], axis=1)
    return df
```
