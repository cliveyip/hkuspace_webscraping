```python
response = requests.post(url, headers=headers, data=payload)
# response = requests.post(url, headers=headers, data=payload)
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
def soup_to_df(soup):
    
    tr_all = soup.select('tbody tr')
    
    Registration_Date = []
    Building_Address = []
    Price = []
    Price_per_Square = []

    for td in tr_all[4:-2]:
        try:
            Registration_Date.append(td.select('td')[1].text.strip())
            Building_Address.append(td.select('td')[2].text.strip().replace("\xa0"," "))
            Price.append(td.select('td')[6].text.strip() + 'M')
            Price_per_Square.append(td.select('td')[7].text.strip())
        except:
            continue

    results_dict = {'Registration_Date': Registration_Date, 'Building_Address': Building_Address, 'Price': Price, 'Price_per_Square': Price_per_Square}

    df = pd.DataFrame(results_dict)
    return df
```

```python
soup_to_df(soup)
```
