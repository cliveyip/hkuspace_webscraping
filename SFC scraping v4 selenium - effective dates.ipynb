```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import re
import pandas as pd
import time

```

```python
df = pd.read_excel('sfc_list.xlsx')
```

```python
cerefs = df['ceref'].to_list()
# cerefs = ['BTT515', 'AAF564', 'AAG749']
```

```python
cerefs
```

```python
# Create a new instance of the Chrome driver
urls = []
eff_dates = []

driver = webdriver.Chrome()

for ref in cerefs:
    url = "https://apps.sfc.hk/publicregWeb/corp/"+ref+"/details"
    urls.append(url)
    print(url)

    # Navigate to the website
    driver.get(url)

    # Sleep for 2 seconds
    time.sleep(2)

    try:
        activity_date_grid = driver.find_element(By.ID, "gridview-1026")
        date = grid_WebElement_to_date(activity_date_grid)
    except:
        date = 'No date found'
    
    eff_dates.append(date)

driver.quit()
```

```python
# add the new columns to the existing df
df['effective_date'] = eff_dates
df['URL'] = urls
df
```

```python
def grid_WebElement_to_date(grid):

    date = ''
    text = grid.text
    
    # Regular expression pattern to match 'Asset Management' followed by a date in 'DD mmm YYYY' format
    pattern = r"Asset Management\n(\d{2} [A-Za-z]{3} \d{4})"

    # Search for the pattern in the input text
    match = re.search(pattern, text)

    # If a match is found, extract the date
    if match:
        date_after_asset_management = match.group(1)
        date = date_after_asset_management
    else:
        date = 'no date found'

    return date
```
