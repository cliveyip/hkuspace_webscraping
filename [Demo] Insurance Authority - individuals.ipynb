```python
import pandas as pd
import requests
```

Part 1 - get the JSON object from the API endpoint

```python
first_name = "Ka Chun"
last_name = "Wong"
```

```python
API_endpoint = "https://iir.ia.org.hk/IISPublicRegisterRestfulAPI/v1/search/individual?seachIndicator=engName&status=A&surNameValue="+last_name+"&givenNameValue="+first_name+"&page=1&pagesize=10"
```

```python
headers = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'}
```

```python
response = requests.get(API_endpoint, headers=headers)
```

```python
response.json()
```

Part 2 - Parse the JSON object for details about the individuals scraped

```python
json = response.json()
```

```python
json['data'][0]
```

```python
json['data'][0]['appointingPrincipalEng']
```

```python
json['itemsCount']
```

```python
def parseJson(json, data_field):
    return json[data_field]
```

```python
licenseNo = []
chiName = []
licenseStatus = []
licenseType = []
appointingPrincipalEng = []
suspendStatus = []
key = []
```

```python
for json in response.json()['data']:
    licenseNo.append(parseJson(json, 'licenseNo'))
    chiName.append(parseJson(json, 'chiName'))
    licenseStatus.append(parseJson(json, 'licenseStatus'))
    licenseType.append(parseJson(json, 'licenseType'))
    appointingPrincipalEng.append(parseJson(json, 'appointingPrincipalEng'))
    suspendStatus.append(parseJson(json, 'suspendStatus'))
```

```python
df = pd.DataFrame({'licenseNo': licenseNo,
                   'chiName': chiName,
                   'licenseStatus': licenseStatus,
                   'licenseType': licenseType,
                   'appointingPrincipalEng': appointingPrincipalEng,
                   'suspendStatus': suspendStatus})
```
