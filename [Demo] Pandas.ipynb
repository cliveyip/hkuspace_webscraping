Create a DataFrame from a List

```python
# lists
OrderDate = ['1/1/2015','1/2/2015','1/5/2015','1/8/2015','1/10/2015']
OrderNumber = ['S0001', 'S0002', 'S0003', 'S0004', 'S0005']
ProductKey = [332, 312, 350, 338, 312]
OrderQuantity = [1, 2, 1, 3, 1]

# create a dictionary from lists
sales_dictionary = {'order_date': OrderDate, 
                    'order_number': OrderNumber, 
                    'product_key': ProductKey,
                    'order_quantity': OrderQuantity}

# create a DataFrame from the dictionary
df = pd.DataFrame(sales_dictionary)

# print the results
df.head()
```
