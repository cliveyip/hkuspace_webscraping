```python
import requests
import pandas as pd
```

```python
# get all the 24 results on the 2nd page
payload = {"frontend_id_list":"10","not_cal_destination":1,"size":24,"start":2,"init_filter":1}

url = 'https://www.klook.com/v1/experiencesrv/search/vertical_service/activity'

headers = {
    'authority': 'www.klook.com',
    'content-type': 'application/json;charset=UTF-8',
    'referer': 'https://www.klook.com/zh-HK/experiences/list/day-trips/cate10/?spm=TNA_Vertical.CategorySeeAll&clickId=be9d84df18',
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/103.0.0.0 Safari/537.36'
}

response = requests.request("POST", url, headers=headers, json=payload)
```
