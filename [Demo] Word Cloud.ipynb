```python
import jieba
# tokenize Chinese words

text = """
本港目前提供31間檢疫酒店供入境人士預訂，惟酒店房間近日供不應求。
有港人原打算本周四（3日）由南韓返港，惟近日發現除了1.6萬元一晚的總統套房，所有檢疫酒店房間已滿，持有商務簽證的他或要滯留南韓。
本報昨瀏覽目前檢疫酒店可供訂房的日子，若入境者要檢疫21日，最快要到本月底才有位，部分酒店更要等到9月，料暑假房間將更「爆棚」。
"""
text_list = list(jieba.cut(text))

wc = WordCloud(font_path=font_path, background_color="white", width=1024, height=768, 
               max_words=100, max_font_size=200, stopwords=stop_words)
wc.generate(" ".join(text_list))

plt.axis("off")
plt.imshow(wc, interpolation="bilinear")
plt.show()
```
