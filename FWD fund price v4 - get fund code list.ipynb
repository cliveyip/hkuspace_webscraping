Step 2: scrape fund price for each fund code

```python
def scrape(fundCode, categoryCode, limit):

    url = 'https://1web.apse1.dgt.api.fwd.com/api/fund/price/historical'

    payload= {"market":"HK","locale":"zh-CN","offset":0,"limit": limit,"fundCode": fundCode,"categoryCode": categoryCode}

    headers={
        "accept": "application/json, text/plain, */*",
        "accept-encoding": "gzip, deflate, br",
        "accept-language": "zh-TW,zh;q=0.9,en-US;q=0.8,en;q=0.7",
        "content-length": "92",
        "origin": "https://www.fwd.com.hk",
        "referer": "https://www.fwd.com.hk/",
        "sec-ch-ua": "\"Chromium\";v=\"116\", \"Not)A;Brand\";v=\"24\", \"Google Chrome\";v=\"116\"",
        "sec-ch-ua-mobile": "?0",
        "sec-ch-ua-platform": "\"Windows\"",
        "sec-fetch-dest": "empty",
        "sec-fetch-mode": "cors",
        "sec-fetch-site": "cross-site",
        "user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/116.0.0.0 Safari/537.36",
    }

    response = requests.post(url, json=payload, headers=headers)

    return response
```

```python
def scrape_full_record(fundCode, categoryCode):
    # first request, scrape 10 items, just to get the total record count
    response = scrape(fundCode, categoryCode, 10)
    totalRecords = response.json()['result']['totalRecords']
    response2 = scrape(fundCode, categoryCode, totalRecords)
    return response2
```

```python
# helper function to transform json response to df
def response2df(response):
    prices = response.json()['result']['prices']
    df = pd.DataFrame(prices)
    # df['fundDescription'] = response.json()['result']['fundDescription']
    df['fundCode'] = response.json()['result']['fundCode']
    df2 = pd.DataFrame(df.price.to_list())
    df3 = df.join(df2)
    df3.drop(columns='price', inplace=True)
    cols = ['fundCode', 'valuationDate', 'bidPrice', 'offerPrice']
    df3 = df3[cols]
    return df3
```

```python
df_all = pd.DataFrame()
for fundCode in fundCode_list:
    print(fundCode)
    response = scrape_full_record(fundCode, categoryCode)
    df = response2df(response)
    df_all = pd.concat([df_all, df])
    time.sleep(1)
    
```
