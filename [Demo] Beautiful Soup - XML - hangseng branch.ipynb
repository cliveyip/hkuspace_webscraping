```python
import requests
from bs4 import BeautifulSoup
```

```python
response = requests.get('https://www.hangseng.com/content/hase/en_hk/configuration/json/banking/branch_list/_jcr_content/upload/file.json?_=1672385329341')
```

```python
response
```

```python
soup = BeautifulSoup(response.text, 'xml')
```

```python
soup
```

```python
sites = soup.find_all('Site')
```

```python
site_code = []
sub_district = []
outlet_type = []
eng_name = []
chi_name = []
eng_add = []
chi_add = []
weekday_hours = []
sat_hours = []

for site in sites:
    site_code.append(site.find('SiteCode').text)
    sub_district.append(site.find('SubDistrict').text)
    outlet_type.append(site.find('OutletType').text)
    eng_name.append(site.find('SiteNameEN').text)
    chi_name.append(site.find('SiteNameZH').text)
    eng_add.append(site.find('SiteAddressEN').text)
    chi_add.append(site.find('SiteAddressZH').text)
    weekday_hours.append(site.find('ServiceHourWeekdays').text)
    sat_hours.append(site.find('ServiceHourSat').text)
```

```python
eng_add
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({'Site Code': site_code,
                   'District': sub_district,
                   'Outlet Type': outlet_type,
                   'English Name': eng_name,
                  'Chinese Name': chi_name,
                   'English Address': eng_add,
                  'Chinese Address': chi_add,
                  'Weekday Hours': weekday_hours,
                  'Saturday Hours': sat_hours})
```

```python
df
```

```python
df.to_excel('HangSeng branches.xlsx', index=False)
```
